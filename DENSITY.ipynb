```python
import numpy as np
import matplotlib.pyplot as plt
a=np.arange(-3,3,0.1)
print(a)
b=np.arange(-2,5,0.1)
print()
print(b)
s,t=np.meshgrid(a,b)

z=s**2+t**2
k=plt.contourf(a,b,z,cmap='Blues')
plt.title('Filled Contours Plot')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.colorbar()
```

```python
import random 
def Rand(start, end, num): 
    res = []   
    for j in range(num): 
        res.append(random.randint(start, end))   
    return res   
num = 10
start = -10
end = 10
x=Rand(start, end, num)
y=Rand(start,end,num)
(a,b)=np.meshgrid(x,y)
z=a**2+b**2
plt.contourf(a,b,z,cmap='Blues')
```

```python
import numpy as np
import matplotlib.pyplot as plt
a=np.linspace(-9,9,19)
b=np.linspace(-9,9,19)
s,t=np.meshgrid(a,b)
print(a)
print()
print(b)
z=np.sqrt(s**2+t**2)
k=plt.contourf(a,b,z)
plt.title('Filled Contours Plot')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.colorbar()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a=np.linspace(-3,3,100)
b=np.linspace(-3,3,100)
s,t=np.meshgrid(a,b)
z=(s**2*t**2)
k=plt.contourf(a,b,z,cmap='plasma')
plt.title('Filled Contours Plot')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.colorbar()
```
